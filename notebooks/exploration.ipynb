Sanity checks run during development.
Quick scripts to validate each stage of the pipeline.

Check 1: Data extraction 

```python
import json

with open("data/lyrics_raw.jsonl", "r", encoding="utf-8") as f:
    songs = [json.loads(line) for line in f]

print(f"Total songs: {len(songs)}")
for artist in ["Taylor Swift", "Pink Floyd", "Kendrick Lamar"]:
    count = sum(1 for s in songs if s["artist"] == artist)
    print(f"  {artist}: {count}")

print("\n--- Sample lyrics (first 500 chars) ---")
print(songs[0]["title"])
print(songs[0]["lyrics"][:500])
```
